Since our dataset is in .pgm format, we need to convert these to .jpg to be suitable for augmentation software.

```python
import os
import cv2
import shutil

def convert_pgm_to_jpg(input_dir, output_dir):
  """
  Converts .pgm images to .jpg in a directory while preserving folder structure.
  """
  for root, dirs, files in os.walk(input_dir):
    for file in files:
      if file.endswith(".pgm"):
        input_path = os.path.join(root, file)
        relative_path = os.path.relpath(root, input_dir)
        output_subdir = os.path.join(output_dir, relative_path)
        os.makedirs(output_subdir, exist_ok=True)
        output_path = os.path.join(output_subdir, file[:-4] + ".jpg")

        img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
        cv2.imwrite(output_path, img, [int(cv2.IMWRITE_JPEG_QUALITY), 100])


# Define input and output directories
input_dir = "/content/att_faces"
output_dir = "/content/att_faces_jpg"

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Convert PGM to JPG and save in the output directory
convert_pgm_to_jpg(input_dir, output_dir)

print("Conversion complete!")
```
